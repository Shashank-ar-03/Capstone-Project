```python
elec_train = elec_df[-200001::-1]
elec_test = elec_df[-200001:-1]
print(elec_test.count())
print(elec_train.count())
```

```python
print(elec_test.describe())
print(elec_train.describe())
```

```python
import re
elec_train["Hour"]= elec_train["Time"].apply(lambda x: (x[:2]))
elec_train["Year"]= elec_train["Date"].apply(lambda x: (x[-4:]))
elec_train["Month"]= elec_train["Date"].apply(lambda x: (re.split(r'/',x)[1]))
elec_new = elec_train.drop(elec_train.columns[[0,1,2]], axis = 1, inplace = True)
elec_train.head()
```

```python
elec_train.dropna(axis = 0, inplace = True)
```

```python
elec_train['GI_cap']=elec_train['Global_intensity'].apply(lambda x: 1 if x>15 else 0)
```

```python
elec_train_dum = pd.get_dummies(elec_train,drop_first=True)
elec_train_dum
```

```python
elec_train_dum.GI_cap[elec_train_dum['GI_cap']==0] = "<15"
elec_train_dum.GI_cap[elec_train_dum['GI_cap']==1] = ">15"
elec_train_dum
```

```python
elec_y = elec_train_dum[['GI_cap']]
elec_X = elec_train_dum.drop(columns =['GI_cap','Global_intensity'])
print(elec_X)
print(elec_y)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression as LR1
X_train, X_test , y_train , y_test = train_test_split(elec_X,elec_y,test_size=.3,random_state=13)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_train)
train_scaled = scaler.transform(X_train)
test_scaled = scaler.transform(X_test)
```

```python
lg_mod = LR1()
lg_mod.fit(train_scaled, y_train)
lg_mod.score(train_scaled, y_train)
```

```python
# Test

y_predict = lg_mod.predict(test_scaled)

print(lg_mod.score(test_scaled, y_test))
```

```python
# Random Forest

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

#create a new rf classifier
rf = RandomForestClassifier()

#create a dictionary of all values we want to test for n_estimators
params_rf = {'n_estimators': [30],'max_features': [23]}

#use gridsearch to test all values for n_estimators
rf_gs = GridSearchCV(rf, params_rf, cv=4)

#fit model to training data
rf_gs.fit(X_train, y_train)
```

```python
#save best model
rf_best = rf_gs.best_estimator_

#check best n_estimators value
print(rf_gs.best_params_)
```

```python
# Test

y_predict = rf_best.predict(X_test)

print(rf_best.score(X_test, y_test))
```
